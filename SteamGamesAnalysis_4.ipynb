## **Feature Engineering & Data Merging**
We will now create the master player-level dataframe.  The primary goal is to calculate the total spending per player and aggregate behavioral features (friends, achievements, reviews).

```python
import ast

# 1. Calculate Total Spending per Player
# The 'library' column is a string representation of a list; convert it to actual lists
purchased_games_exp = purchased_games.copy()
purchased_games_exp['library'] = purchased_games_exp['library'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])

# Explode the library so each row is a player-game pair
purchased_games_exp = purchased_games_exp.explode('library').rename(columns={'library': 'gameid'})

# Merge with prices to get the cost of each game
# Note: Using the most recent price available in the 'prices' table
latest_prices = prices.sort_values('date_acquired').groupby('gameid').tail(1)
spend_df = purchased_games_exp.merge(latest_prices[['gameid', 'usd']], on='gameid', how='left')

# Aggregate to player level
player_spending = spend_df.groupby('playerid')['usd'].sum().reset_index().rename(columns={'usd': 'total_amount_spent'})

# 2. Aggregate other features
# Friend count
friends_count = friends.copy()
friends_count['friend_count'] = friends_count['friends'].apply(lambda x: len(ast.literal_eval(x)) if isinstance(x, str) else 0)

# Review count
review_stats = reviews.groupby('playerid').agg(num_reviews=('reviewid', 'count'), total_helpful_received=('helpful', 'sum')).reset_index()

# Achievement count
achievement_stats = history.groupby('playerid').agg(total_achievements=('achievementid', 'count')).reset_index()

# 3. Final Merge
final_df = players.merge(player_spending, on='playerid', how='left') \
                  .merge(friends_count[['playerid', 'friend_count']], on='playerid', how='left') \
                  .merge(review_stats, on='playerid', how='left') \
                  .merge(achievement_stats, on='playerid', how='left')

# Fill NaNs for players with no recorded activity in certain tables
final_df[['total_amount_spent', 'friend_count', 'num_reviews', 'total_helpful_received', 'total_achievements']] = \
    final_df[['total_amount_spent', 'friend_count', 'num_reviews', 'total_helpful_received', 'total_achievements']].fillna(0)

display(final_df.head())
print(f"Final shape: {final_df.shape}")
```

### Advanced Feature Engineering
We will now refine the features by adding account age, library diversity, and engagement metrics.

```python
steam_df = final_df.copy()
# 1. Account Age (Years)
steam_df['created'] = pd.to_datetime(steam_df['created'])
reference_date = pd.to_datetime('2025-02-24') # Based on the dataset's latest price date
steam_df['account_age_years'] = (reference_date - steam_df['created']).dt.days / 365.25

# 2. Game Diversity (Unique Games Owned)
game_diversity = purchased_games_exp.groupby('playerid')['gameid'].nunique().reset_index().rename(columns={'gameid': 'unique_games_owned'})

# 3. Review Quality (Sum of Helpful, Funny, Awards)
# Assuming 'funny' and 'awards' columns exist based on reviews.head() output
review_quality = reviews.groupby('playerid').agg(total_social_feedback=pd.NamedAgg(column='helpful', aggfunc='sum')).reset_index()
# We can add 'funny' and 'awards' if we verify they are numeric
review_quality['total_social_feedback'] += reviews.groupby('playerid')[['funny', 'awards']].sum().sum(axis=1).values

# Merge new features
steam_df = steam_df.merge(game_diversity, on='playerid', how='left') \
                  .merge(review_quality, on='playerid', how='left')

# 4. Derived Ratios
steam_df['unique_games_owned'] = steam_df['unique_games_owned'].fillna(0)
steam_df['total_social_feedback'] = steam_df['total_social_feedback'].fillna(0)

# Achievements per game (Engagement Intensity)
steam_df['achievements_per_game'] = steam_df['total_achievements'] / (steam_df['unique_games_owned'] + 1)

display(steam_df[['playerid', 'account_age_years', 'unique_games_owned', 'total_social_feedback', 'achievements_per_game']].head())
print(f"New shape with advanced features: {steam_df.shape}")
```

## **Additional EDA: Engagement vs Spending**

In this section, we examine whether user engagement features such as friend count, achievements, reviews, and games owned are associated with player spending behavior.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

eda_df = steam_df.copy()

# Log transform spending because the raw distribution is highly skewed
eda_df["log_spending"] = np.log1p(eda_df["total_amount_spent"])

features = [
    "friend_count",
    "total_achievements",
    "num_reviews",
    "total_social_feedback",
    "unique_games_owned"
]

sample_df = eda_df.sample(min(5000, len(eda_df)), random_state=42)

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.scatterplot(
        data=sample_df,
        x=feature,
        y="log_spending",
        alpha=0.25,
        ax=axes[i]
    )
    axes[i].set_title(f"{feature} vs Log Spending")
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel("log(1 + total_amount_spent)")

axes[-1].axis("off")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

bucket_df = steam_df.copy()

bucket_df["friend_bucket"] = pd.qcut(
    bucket_df["friend_count"].rank(method="first"),
    q=4,
    labels=["Low", "Mid-Low", "Mid-High", "High"]
)

bucket_df["achievement_bucket"] = pd.qcut(
    bucket_df["total_achievements"].rank(method="first"),
    q=4,
    labels=["Low", "Mid-Low", "Mid-High", "High"]
)

bucket_df["review_bucket"] = pd.qcut(
    bucket_df["num_reviews"].rank(method="first"),
    q=4,
    labels=["Low", "Mid-Low", "Mid-High", "High"]
)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

friend_summary = bucket_df.groupby("friend_bucket", observed=True)["total_amount_spent"].mean().reset_index()
achievement_summary = bucket_df.groupby("achievement_bucket", observed=True)["total_amount_spent"].mean().reset_index()
review_summary = bucket_df.groupby("review_bucket", observed=True)["total_amount_spent"].mean().reset_index()

sns.barplot(data=friend_summary, x="friend_bucket", y="total_amount_spent", ax=axes[0])
axes[0].set_title("Average Spending by Friend Count Bucket")
axes[0].set_xlabel("Friend Count Bucket")
axes[0].set_ylabel("Average Total Amount Spent (USD)")

sns.barplot(data=achievement_summary, x="achievement_bucket", y="total_amount_spent", ax=axes[1])
axes[1].set_title("Average Spending by Achievement Bucket")
axes[1].set_xlabel("Achievement Bucket")
axes[1].set_ylabel("Average Total Amount Spent (USD)")

sns.barplot(data=review_summary, x="review_bucket", y="total_amount_spent", ax=axes[2])
axes[2].set_title("Average Spending by Review Count Bucket")
axes[2].set_xlabel("Review Count Bucket")
axes[2].set_ylabel("Average Total Amount Spent (USD)")

plt.tight_layout()
plt.show()
```

The additional EDA suggests that player engagement is positively associated with spending behavior, although the relationship is not equally strong across all features. In the scatterplots, higher values of engagement variables such as friend count, number of reviews, total achievements, and unique games owned generally correspond to higher spending, but the points are widely spread, showing that the relationship is noisy rather than perfectly linear.

The bucketed bar charts make this pattern easier to see. On average, players in higher engagement buckets tend to spend more than players in lower engagement buckets. This trend is most noticeable for review count and achievements, which suggests that players who are more active on the platform also tend to purchase more. The strongest association appears for unique games owned, which is expected because owning more games is directly tied to making more purchases.

Overall, these visualizations provide preliminary support for our hypothesis that more engaged users tend to spend more on Steam. However, since the patterns are not perfectly consistent and spending is highly skewed, these plots should be interpreted as descriptive evidence rather than proof of a statistically significant relationship. Formal hypothesis testing and regression analysis would be needed to confirm the strength and significance of these effects.

```python
import numpy as np
import pandas as pd

df = steam_df.copy()

df = df.replace([np.inf, -np.inf], np.nan)

df_model = df.dropna(subset=[
    'unique_games_owned',
    'friend_count',
    'account_age_years',
    'total_amount_spent'
])
```

```python
df_model['log_spending'] = np.log1p(df_model['total_amount_spent'])
df_model['log_games'] = np.log1p(df_model['unique_games_owned'])
```

```python
df_model['game_bucket'] = pd.cut(
    df_model['unique_games_owned'],
    bins=[-1, 0, 10, 50, df_model['unique_games_owned'].max()],
    labels=['0 games', '1-10', '11-50', '51+']
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.boxplot(
    x='game_bucket',
    y='log_spending',
    data=df_model
)

plt.title('Log Spending by Game Ownership Group')
plt.xlabel('Game Ownership Bucket')
plt.ylabel('log(1 + Spending)')
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.regplot(
    x='unique_games_owned',
    y='total_amount_spent',
    data=df_model,
    scatter_kws={'alpha':0.1}
)

plt.title('Games Owned vs Spending')
plt.show()
```

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(
    df_model['unique_games_owned'],
    df_model['total_amount_spent']
)

print("Correlation:", corr)
print("p-value:", p_value)
```

```python
from scipy.stats import ttest_ind

median = df_model['unique_games_owned'].median()

low_group = df_model[df_model['unique_games_owned'] <= median]['total_amount_spent']
high_group = df_model[df_model['unique_games_owned'] > median]['total_amount_spent']

t_stat, p_value = ttest_ind(high_group, low_group, equal_var=False)

print("T-stat:", t_stat)
print("p-value:", p_value)
```

```python
import statsmodels.api as sm

X = df_model[['unique_games_owned', 'friend_count', 'account_age_years']]
X = sm.add_constant(X)

y = df_model['total_amount_spent']

model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
X = df_model[['log_games', 'friend_count', 'account_age_years']]
X = sm.add_constant(X)

y = df_model['log_spending']

model_log = sm.OLS(y, X).fit()

print(model_log.summary())
```

The hypothesis test strongly supports the relationship between game ownership and player spending. The Pearson correlation between unique games owned and total spending was very high and statistically significant, r = 0.943, p < 0.001. A two-sample t-test also showed that players with above-median game ownership spent significantly more than those with below-median ownership, t = 73.93, p < 0.001. In the OLS model, unique games owned remained a strong predictor of total spending after controlling for friend count and account age. The raw spending model explained 88.9% of the variation in spending, while the log-transformed model explained 97.9%. Overall, these findings support the hypothesis that players with larger game libraries tend to spend substantially more on Steam.
